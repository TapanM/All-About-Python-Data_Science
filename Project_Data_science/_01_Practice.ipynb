# WebScrapping Practice

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Demographics_of_India"

web_data = requests.get(url).text
```

```python
soup = BeautifulSoup(web_data, "html5lib")
print(soup.prettify)
```

```python
tables = soup.findAll("table")
len(tables)
```

```python
for index, table in enumerate(tables):
    if("Population growth of India per decade" in str(table)):
        table_index = index
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
demographic_data = pd.DataFrame(
    columns=["Census year", "Population", "Change(%)"])

for row in tables[table_index].find("tbody").find_all("tr"):
    col = row.find_all("td")
    if(col != []):
        census_year = col[0].text.strip()
        population = col[1].text.strip()
        change = col[2].text.strip()
        demographic_data = demographic_data.append({"Census year": census_year, "Population": population, "Change(%)": change}, ignore_index=True)
```
